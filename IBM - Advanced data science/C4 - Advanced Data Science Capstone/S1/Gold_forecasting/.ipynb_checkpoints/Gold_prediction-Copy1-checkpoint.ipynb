```python
#import packages
import numpy as np
import pandas as pd
from keras.preprocessing import sequence
from keras.models import load_model
```

```python
# defining the batch size and number of epochs
batch_size = 32
epochs = 200
timesteps = 20
```

```python
def get_train_length(dataset, batch_size, test_percent):
    # substract test_percent to be excluded from training, reserved for testset
    length = len(dataset)
    length *= 1 - test_percent
    train_length_values = []
    for x in range(int(length) - batch_size*4,int(length)): 
        modulo=x%batch_size
        if (modulo == 0):
            train_length_values.append(x)
            print(x)
    return (max(train_length_values))
```

```python
def get_test_length(dataset, batch_size):
    
    test_length_values = []
    for x in range(len(dataset) - 200, len(dataset) - timesteps*2): 
        modulo=(x-upper_train)%batch_size
        if (modulo == 0):
            test_length_values.append(x)
            print(x)
    return (max(test_length_values))
```

```python
length = get_train_length(df_gold, batch_size, 0.20)
print(length)
```

```python
#Adding timesteps * 2
upper_train = length + timesteps*2
df_gold_train = df_gold[0:upper_train]
training_set = df_gold_train.iloc[:,1:2].values
training_set.shape
```

```python
test_length = get_test_length(df_gold, batch_size)
print(test_length)
upper_test = test_length + timesteps*2
testset_length = test_length - upper_train
print(testset_length)
```

```python
#load model
import h5py
model = load_model(filepath="gold_model_with_sequential_test.h5")
```

```python
df_gold.head(2)
```

```python
samples = upper_test - upper_train
samples
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range = (0, 1))
# construct test set

#subsetting
df_data_1_test = df_gold.iloc[-samples: , :] 
test_set = df_data_1_test.iloc[:,1:2].values

#Preprocessing data set
test_set_norm = np.array(test_set).reshape(-1,1)

#scaling
scaled_real_bcg_values_test = sc.fit_transform(np.float64(test_set_norm))
```

```python
scaled_real_bcg_values_test.shape
```

```python
#creating input data
X_test = []
for i in range(timesteps, testset_length + timesteps):
    X_test.append(scaled_real_bcg_values_test[i-timesteps:i, 0])
X_test = np.array(X_test)


#reshaping
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
```

```python
X_test.shape
```

```python
#prediction
predicted_bcg_values_test_mae = model.predict(X_test, batch_size=batch_size)
print(predicted_bcg_values_test_mae.shape)

#reshaping
predicted_bcg_values_test_mae = np.reshape(predicted_bcg_values_test_mae, 
                                       (predicted_bcg_values_test_mae.shape[0], 
                                        predicted_bcg_values_test_mae.shape[1]))

print(predicted_bcg_values_test_mae.shape)

#inverse transform
predicted_bcg_values_test_mae = sc.inverse_transform(predicted_bcg_values_test_mae)


```

```python
predicted_bcg_values_test_mae[959, ]
```

```python
#creating y_test data
y_test = []
for i in range(0, 49):
    for j in range(0,20):
        if i > 0:
            y_test = np.append(y_test, predicted_bcg_values_test_mae[(i*20)-1, j])
        else:
            y_test = np.append(y_test, predicted_bcg_values_test_mae[(i*20), j])
                

# reshaping
y_test = np.reshape(y_test, (y_test.shape[0], 1))

print(y_test.shape)
```

```python
predicted_bcg_values_test_mae[timesteps-timesteps+1, (timesteps-1)-1]
```

```python
#creating y_test data
y_test = []
for i in range(timesteps, testset_length-timesteps):
    total = 0
    for j in range(0,timesteps):
        total = total + predicted_bcg_values_test_mae[i-timesteps+j, (timesteps-1)-j]
        #v0 = predicted_bcg_values_test_mae[i-timesteps, (timesteps-1)]
        #v1 = predicted_bcg_values_test_mae[i-timesteps+1, (timesteps-1)-1]
        #v2 = predicted_bcg_values_test_mae[i-timesteps+2, (timesteps-1)-2]
    mean = 0 
    mean = total/timesteps
    y_test = np.append(y_test, mean)
                

# reshaping
y_test = np.reshape(y_test, (y_test.shape[0], 1))

print(y_test.shape)
```

```python
#subsetting
df_data_1_test = df_gold[-samples:] 
test_date_set = df_data_1_test.iloc[:,0:2]
test_date_set.reset_index(inplace=True, drop=True)
```

```python
#creating y_test data
y_test = []
for i in range(0, testset_length):
    total = 0
    mean = 0 
    
    if i >=timesteps and i < (testset_length-timesteps):
        for j in range(0,timesteps):
            total = total + predicted_bcg_values_test_mae[i-timesteps+j, (timesteps-1)-j]
            #v0 = predicted_bcg_values_test_mae[i-timesteps, (timesteps-1)]
            #v1 = predicted_bcg_values_test_mae[i-timesteps+1, (timesteps-1)-1]
            #v2 = predicted_bcg_values_test_mae[i-timesteps+2, (timesteps-1)-2]

        mean = total/timesteps
        y_test = np.append(y_test, mean)
        
        
    if i < timesteps :
        for j in range(0,i+1):
            total = total + predicted_bcg_values_test_mae[i-j, j]
            
        mean = total/(i+1)
        y_test = np.append(y_test, mean)
        
    if i >= (testset_length-timesteps):
        i_aux = i - (testset_length-timesteps)
        for j in range(0,i_aux+1):
            total = total + predicted_bcg_values_test_mae[i-j, j]
            
        mean = total/(i_aux+1)
        y_test = np.append(y_test, mean)
        
                

# reshaping
y_test = np.reshape(y_test, (y_test.shape[0], 1))

print(y_test.shape)
```
