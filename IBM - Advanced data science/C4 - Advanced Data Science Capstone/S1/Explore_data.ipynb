```python
df_euro_currency =  pd.read_csv("euro-daily-hist_1999_2020.csv")
df_euro_currency.head(2)
```

```python
df_euro_currency.columns = df_euro_currency.columns.str.replace('[', '')
df_euro_currency.columns = df_euro_currency.columns.str.replace(']', '')
df_euro_currency.columns = df_euro_currency.columns.str.rstrip()
df_euro_currency.head(2)
```

```python
df_euro_currency = df_euro_currency.rename(columns={df_euro_currency.columns[0]: 'Date'})
df_euro_currency.head(2)
```

```python
currencies = ['Date', 'Brazilian real', 'Chinese yuan renminbi',  'US dollar', 'UK pound sterling', 'Russian rouble']
df_euro_currency = df_euro_currency[currencies]
df_euro_currency.head(2)
```

```python
columns_to_keep = ['Date', 'Close']
df_gold = df_gold[columns_to_keep] 
df_cruid_oil = df_cruid_oil[columns_to_keep]
df_gas = df_gas[columns_to_keep]
df_heating_oil = df_heating_oil[columns_to_keep]
df_gasoline = df_gasoline[columns_to_keep]
```

```python
new_df = pd.merge(df_gold, df_cruid_oil, on='Date', how='inner')
```

```python
new_df.head()
```

```python
new_df = pd.merge(new_df, df_gas, on='Date', how='inner')
new_df.head()
```

```python
new_df = pd.merge(new_df, df_heating_oil, on='Date', how='inner')
new_df.head()
```

```python
new_df = pd.merge(new_df, df_gasoline, on='Date', how='inner')
new_df.head()
```

```python
new_df = pd.merge(new_df, df_br_cruid_oil, on='Date', how='inner')
new_df.head()
```

```python
new_df.columns = ['Date', 'Gold', 'Cruid Oil', 'Natural Gas', 'Heat Oil', 'Gasoline', 'Brent Oil']
new_df.head()
```

```python
df_euro_currency.head(2)
```

```python
new_df = pd.merge(new_df, df_euro_currency, on='Date', how='inner')
new_df.head()
```

```python
new_df.count()
```

```python
new_df.dtypes
```

```python
new_df[['Brazilian real', 'Chinese yuan renminbi', 'US dollar', 
     'UK pound sterling', 'Russian rouble' ]] = new_df[['Brazilian real', 'Chinese yuan renminbi', 'US dollar',
                                       'UK pound sterling', 'Russian rouble']].apply(pd.to_numeric, errors='coerce')
```

```python
print(new_df.isnull().sum())
```

```python
new_df = new_df.dropna()    # Dropping the missing values.
print(new_df.isnull().sum())
```

```python
new_df.count()
```

```python
for column in  new_df.columns.tolist()[7:]:
    new_df[column] = 1/new_df[column]
```

```python
new_df.head(2)
```

Since our currencies values are mesured in euros, we have to convert the Gold, Oil and Gas prices to euro as well,
we just have to use de euro-dolar conversion and apply it to the firsts columns.
So for each row we just have to divide each column by the Us dollar column

```python
for column in  new_df.columns.tolist()[1:7]:
    new_df[column] = new_df[column]/new_df['US dollar']
```
