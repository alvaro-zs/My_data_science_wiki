```python
import pandas as pd
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
```

```python
previsores = pd.read_csv('entradas_breast.csv')
previsores.head(3)
```

```python
classe = pd.read_csv('saidas_breast.csv')
classe.head(3)
```

```python
def criarRedecomDropout():
    classificador = Sequential()
    classificador.add(Dense(units = 8, activation = 'relu', 
                        kernel_initializer = 'random_uniform', input_dim = 30))
    
    # Dropout evita overfitting - ideal entre 0.2 e 0.3, valores maiores ocasinam underfitting
    classificador.add(Dropout(0.2))
    
    classificador.add(Dense(units = 8, activation = 'relu', 
                        kernel_initializer = 'random_uniform'))
    
    classificador.add(Dropout(0.2))
    
    classificador.add(Dense(units = 1, activation = 'sigmoid'))
    
    otimizador = keras.optimizers.Adam(lr = 0.001, decay = 0.0001, clipvalue = 0.5)
    
    classificador.compile(optimizer = otimizador, loss = 'binary_crossentropy',
                      metrics = ['binary_accuracy'])
    return classificador
```

```python
def criarRedesemDropout():
    classificador = Sequential()
    classificador.add(Dense(units = 16, activation = 'relu', 
                        kernel_initializer = 'random_uniform', input_dim = 30))
    
    
    classificador.add(Dense(units = 16, activation = 'relu', 
                        kernel_initializer = 'random_uniform'))
    
    
    classificador.add(Dense(units = 1, activation = 'sigmoid'))
    
    otimizador = keras.optimizers.Adam(lr = 0.001, decay = 0.0001, clipvalue = 0.5)
    
    classificador.compile(optimizer = otimizador, loss = 'binary_crossentropy',
                      metrics = ['binary_accuracy'])
    return classificador
```

```python
classificador = KerasClassifier(build_fn = criarRedecomDropout,
                                epochs = 100,
                                batch_size = 10)
classificador
```

```python
resultados_Drop = cross_val_score(estimator = classificador,
                             X = previsores, y = classe,
                             cv = 10, scoring = 'accuracy')
```

```python
classificador = KerasClassifier(build_fn = criarRedesemDropout,
                                epochs = 100,
                                batch_size = 10)
classificador
```

```python
resultados_sem_Drop = cross_val_score(estimator = classificador,
                             X = previsores, y = classe,
                             cv = 10, scoring = 'accuracy')
```

```python
resultados_sem_Drop
```

```python
resultados_Drop
```

```python
media = resultados_sem_Drop.mean()
print('accuracy mean SEM Dropout: {}'.format(media))
media = resultados_Drop.mean()
print('accuracy mean COM Dropout: {}'.format(media))
```

```python
desvio = resultados_sem_Drop.std()
print('Desvio SEM Dropout: {}'.format(desvio))
desvio = resultados_Drop.std()
print('Desvio COM Dropout: {}'.format(desvio))
```
